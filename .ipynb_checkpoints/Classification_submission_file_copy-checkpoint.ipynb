### Distribution of word count

```python
def wordcount(data):
    """
    Returns a graph showing distribution of word count for each sentiment

     Parameters
    -----------
    data: DataFrame.
    
    """
    
    word_count=data['message'].apply(lambda x: len(x.split()))
    wordcount_sentiment=pd.concat([pd.DataFrame(word_count),data['sentiment']],axis=1)
    ax = sns.catplot(x="sentiment", y="message",kind='boxen', data=wordcount_sentiment)
    plt.ylabel('Word count', fontsize=12)
```

### Extracting hashtags and Twitter handes

In the quest to find out what makes each class unique we can try to explore what topics each class talks about the most, we can do this by extracting the hashtags from the tweets and creating plots of the most mentioned hashtags for each sentiment group. We can also find out who or what entity is being tagged the most in each sentiment group we do this by extraching the mentioned twitter handles from the tweets.

By using the regular expression library we can then extract the 2 features and plot the top 5 most frequent topics and twitter accounts.

```python
def get_hashtag(data):
    """
    Returns hashtags from our tweets.

     Parameters
    -----------
    data: DataFrame column that contains tweets.
    
    """
    hashtags=[]
    for message in data:
        #converting tweet to lowercase
        message=message.lower()
        #Finding all hashtages
        message=re.findall('#([^\s]+)',message)
        hashtags.append(message)
    hash_list=[tag for tag in hashtags if tag != []]
    list_df=pd.DataFrame(hash_list)
    hashtags=list_df.stack(level=0)
    
    return hashtags
```

```python
def get_handle(data):
    """
    Returns twitter handles from our tweets.

     Parameters
    -----------
    data: DataFrame column that contains tweets.
   
    """
    handles=[]
    for message in data:
        #Finding all user
        message=re.findall('@([^\s]+)',message)
        handles.append(message)
        
    user_list=[user for user in handles if user != []]
    list_df=pd.DataFrame(user_list)
    handles=list_df.stack(level=0)
    return handles
```

# Feature Engineering

```python
#Function that Preprocess tweets
def data_preprocessing(data): 
    """
    Returns preprocessed tweets and the function  removes stopwords,links,usernames.hashtags,words containing digits,
    non alphabetical words,lemmatizes text and also removes words with length less tha 3 
    
     Parameters
    -----------
    data: DataFrame column that contains tweet data.
    
    """
    stop_words = set(stopwords.words('english')) 
    cleaned_tweet=[]
    for message in data:
        #converting tweets to lower case
        message=message.lower()
        
        #Deleting Links from tweets
        message = re.sub(r"http\S+",'',message)
    
        #Removing username from tweets
        message = re.sub('@[^\s]+','',message)
        
        #Removing hashtags on tweets
        message = re.sub('#([^\s]+)','', message)
        
        #Removing stop words
        word_tokens = word_tokenize(message) 
        message = [word for word in word_tokens if not word in stop_words]
        message = " ".join(message)
        
        #Remove words that contain digits
        splitted_word = message.split()
        message = [word for word in splitted_word if not any(c.isdigit() for c in word)]
        message = " ".join(message)
        
        #Replace non_aphabets with space
        message = re.sub('[^a-zA-Z0-9]',' ', message)
        
        #Removing words with length <=3
        shortword = re.compile(r'\W*\b\w{1,3}\b')
        message=shortword.sub('',message)
        
        #removing repeated letters of more than two in a words. e.g greeaaatt -> great
        message=re.sub(r'(.)\1+', r'\1\1',message)
        
        #Lemmatizer 
        wordLemm = WordNetLemmatizer()
        lemmatize_word=''
        for word in message.split():
            word=wordLemm.lemmatize(word)
            lemmatize_word=lemmatize_word+(word+' ')
            message=lemmatize_word.lstrip()
            
        cleaned_tweet.append(message)
        
    return cleaned_tweet
```

```python
#A function that returns non stopwords
def not_stopword(data):
    
    """
    Returns words that are not stopwords.
    
     Parameters
    -----------
    
    data: DataFrame column that contains tweets.
    """
    output=[]
    for message in data:
        stop_words = set(stopwords.words('english'))
        message = re.sub('[^a-zA-Z0-9]',' ', message)
        message=[word.lower() for word in message.split() if word.lower() not in stop_words]
        output.append(message)
    stop_list=[lists for lists in output if lists != []]
    list_df=pd.DataFrame(stop_list)
    list_df_stack=list_df.stack(level=0)
    return list_df_stack
```

```python
def most_common_words(anti_data,neutral_data,pro_data,news_data): 
    """
    Returns graphs of top 10 words for each sentiment on our tweets data.
    
     Parameters
    -----------
    anti_data: DataFrame column that sentiment of -1.
    neutral_data: DataFrame column that sentiment of 0.
    pro_data: DataFrame column that sentiment of 1.
    news_data: DataFrame column that sentiment of 2.
    """
    fig,ax = plt.subplots(2,2,figsize=(11, 10))
    fig.tight_layout(pad=7.0)
    sns.countplot(y=not_stopword(anti_data), order = not_stopword(anti_data).value_counts()[0:10].index,ax=ax[0,0],color='cadetblue')
    sns.countplot(y=not_stopword(neutral_data), order = not_stopword(neutral_data).value_counts()[0:10].index,ax=ax[0,1],color='cadetblue')
    sns.countplot(y=not_stopword(pro_data), order = not_stopword(pro_data).value_counts()[0:10].index,ax=ax[1,0],color='cadetblue')
    sns.countplot(y=not_stopword(news_data), order = not_stopword(news_data).value_counts()[0:10].index,ax=ax[1,1],color='cadetblue')
    sns.color_palette("BuGn_r")
    ax[0,0].set_title('Most common words (sentiment=-1)',fontsize=11)
    ax[0,1].set_title('Most common words (sentiment=0)',fontsize=11)
    ax[1,0].set_title('Most common words (sentiment=1)',fontsize=11)
    ax[1,1].set_title('Most common words (sentiment=2)',fontsize=11)
    plt.show()
```

Function for ngrams

```python
def get_ngram(data, n=None):
    
    """
    Returns n_grams from tweets.
    
     Parameters
    -----------
    data: DataFrame column that contains tweets.
    n: Number for n_grams
    """
    vec = CountVectorizer(ngram_range=(n, n)).fit(data)
    bag_of_words = vec.transform(data)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) 
                    for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:10]
```

```python
def n_gram_plot(anti_data,neutral_data,pro_data,news_data,n_gram):
        
    """
    Returns a graph showing n_grams for each sentiment.
    
     Parameters
    -----------
    sentiments:anti_data,neutral_data,pro_data,news_data
    n_gram: Number for n_grams
    """
    fig,ax = plt.subplots(2,2,figsize=(11, 10))
    fig.tight_layout(pad=9.0)
    x_anti,y_anti=map(list,zip(*get_ngram(anti_data, n_gram)))
    x_nue,y_nue=map(list,zip(*get_ngram(neutral_data, n_gram)))
    x_pro,y_pro=map(list,zip(*get_ngram(pro_data, n_gram)))
    x_news,y_news=map(list,zip(*get_ngram(news_data, n_gram)))
    sns.barplot(x=y_anti,y=x_anti,ax=ax[0,0],color='cadetblue')
    sns.barplot(x=y_nue,y=x_nue,ax=ax[0,1],color='cadetblue')
    sns.barplot(x=y_pro,y=x_pro,ax=ax[1,0],color='cadetblue')
    sns.barplot(x=y_news,y=x_news,ax=ax[1,1],color='cadetblue')
    ax[0,0].set_title('(sentiment=-1)',fontsize=11)
    ax[0,1].set_title('(sentiment=0)',fontsize=11)
    ax[1,0].set_title('(sentiment=1)',fontsize=11)
    ax[1,1].set_title('(sentiment=2)',fontsize=11)
    plt.show()
```

```python
def name_entity(data):
    
    """
    Returns  name entity from data.
    
     Parameters
    -----------
    data:Dataframe containing tweets data column
    """
    nlp = en_core_web_md.load()
    entites_output=[]
    for message in data:
        doc=nlp(message)
        entities=[word.label_ for word in doc.ents]
        entites_output.append(entities)
        
    entity_list=[ent for ent in entites_output if ent != []]
    entity_df=pd.DataFrame(entity_list)
    entity=entity_df.stack(level=0)
    return entity
```

#### Top 5 Part of Speech  in our tweet data

```python
def parts_of_speach(text):
    
    """
    Returns  Part of speech from data.
    
     Parameters
    -----------
    data:Dataframe containing tweets data column
    """
    
    def _get_pos(text):
        pos=nltk.pos_tag(word_tokenize(text))
        pos=list(map(list,zip(*pos)))[1]
        return pos
    
    tags=text.apply(lambda x : _get_pos(x))
    tags=[x for l in tags for x in l]
    counter=Counter(tags)
    
    return counter
```

```python
def plot_Part_of_speech(anti_tweet,neutral_tweet,pro_tweet,news_tweet):
    
    """
    Returns a graph showing Part of speech count on each sentiment.
    
     Parameters
    -----------
    sentiments:anti_data,neutral_data,pro_data,news_data
    """
    
    fig,ax = plt.subplots(2,2,figsize=(8,6))
    fig.tight_layout(pad=5)
    x_anti,y_anti=map(list,zip(*parts_of_speach(anti_tweet).most_common(5)))
    x_nue,y_nue=map(list,zip(*parts_of_speach(neutral_tweet).most_common(5)))
    x_pro,y_pro=map(list,zip(*parts_of_speach(pro_tweet).most_common(5)))
    x_news,y_news=map(list,zip(*parts_of_speach(news_tweet).most_common(5)))
    sns.barplot(x=y_anti,y=x_anti,ax=ax[0,0],color='cadetblue')
    sns.barplot(x=y_nue,y=x_nue,ax=ax[0,1],color='cadetblue')
    sns.barplot(x=y_pro,y=x_pro,ax=ax[1,0],color='cadetblue')
    sns.barplot(x=y_news,y=x_news,ax=ax[1,1],color='cadetblue')
    ax[0,0].set_title('Part of speech (sentiment=-1)',fontsize=11)
    ax[0,1].set_title('Part of speech  (sentiment=0)',fontsize=11)
    ax[1,0].set_title('Part of speech  (sentiment=1)',fontsize=11)
    ax[1,1].set_title('Part of speech  (sentiment=2)',fontsize=11)
    plt.show()
```
