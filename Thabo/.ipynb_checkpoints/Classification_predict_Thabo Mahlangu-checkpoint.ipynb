## 4. Tweets Preprocessing and EDA

#### 4.1 Displaying Text features before data_preprocessing

***Function to display text features***

```python
#Function to get hashtags and tweets
def get_hashtag(data):
    hashtags=[]
    for message in data:
        #converting tweet to lowercase
        message=message.lower()
        #Finding all hashtages
        message=re.findall('#([^\s]+)',message)
        hashtags.append(message)
    hash_list=[tag for tag in hashtags if tag != []]
    list_df=pd.DataFrame(hash_list)
    hashtags=list_df.stack(level=0)
    
    return hashtags

def get_handle(data):
    handles=[]
    for message in data:
        #Finding all user
        message=re.findall('@([^\s]+)',message)
        handles.append(message)
        
    user_list=[user for user in handles if user != []]
    list_df=pd.DataFrame(user_list)
    handles=list_df.stack(level=0)
    return handles

def get_url(data):
    urls=[]
    for message in data:
        #Finding all user
        message=re.findall(r"http\S+",message)
        urls.append(message)
    return urls

def get_stopword(data):
    output=[]
    for message in data:
        stop_words = set(stopwords.words('english')) 
        message=[word.lower() for word in message.split() if word.lower() in stop_words]
        output.append(message)
    stop_list=[lists for lists in output if lists != []]
    list_df=pd.DataFrame(stop_list)
    list_df_stack=list_df.stack(level=0)
    return list_df_stack

def not_stopword(data):
    output=[]
    for message in data:
        stop_words = set(stopwords.words('english'))
        message = re.sub('[^a-zA-Z0-9]',' ', message)
        message=[word.lower() for word in message.split() if word.lower() not in stop_words]
        output.append(message)
    stop_list=[lists for lists in output if lists != []]
    list_df=pd.DataFrame(stop_list)
    list_df_stack=list_df.stack(level=0)
    return list_df_stack


def get_ngram(data, n=None):
    vec = CountVectorizer(ngram_range=(n, n)).fit(data)
    bag_of_words = vec.transform(data)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) 
                    for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:10]

def name_entity(data):
    nlp = en_core_web_sm.load()
    entites_output=[]
    for message in data:
        doc=nlp(message)
        entities=[word.label_ for word in doc.ents]
        entites_output.append(entities)
        
    entity_list=[ent for ent in entites_output if ent != []]
    entity_df=pd.DataFrame(entity_list)
    entity=entity_df.stack(level=0)
    return entity


def name_entity_label(data):
    nlp = en_core_web_sm.load()
    entites_output=[]
    for message in data:
        doc=nlp(message)
        entities=[word.text for word in doc.ents if word.label_ =='ORG']
        entites_output.append(entities)
        
    entity_list=[ent for ent in entites_output if ent != []]
    entity_df=pd.DataFrame(entity_list)
    entity=entity_df.stack(level=0)
    return entity


    
def parts_of_speach(text):
    
    def _get_pos(text):
        pos=nltk.pos_tag(word_tokenize(text))
        pos=list(map(list,zip(*pos)))[1]
        return pos
    
    tags=text.apply(lambda x : _get_pos(x))
    tags=[x for l in tags for x in l]
    counter=Counter(tags)
    
    return counter
  
```

### 4.2 Text_Preprocessing (Cleaning text)

```python
#Function that Preprocess tweets
def data_preprocessing(data):
    stop_words = set(stopwords.words('english')) 
    results=[]
    for message in data:
        #converting tweets to lower case
        message=message.lower()
        
        #Deleting Links from tweets
        message = re.sub(r"http\S+",'',message)
        
        #Removing username from tweets
        message = re.sub('@[^\s]+','',message)
        
        #Removing hashtags on tweets
        message = re.sub('#([^\s]+)','', message)
        
        #Replacing shortwords with full words
        splitted_word = message.split()
        message = [Short_words[x] if x in Short_words else x for x in splitted_word]
        message = " ".join(message)
        
        #Replacing Abbreviations
        splitted_word = message.split()
        message = [abbreviations[x] if x in abbreviations else x for x in splitted_word]
        message = " ".join(message)
        
        #Removing stop words
        word_tokens = word_tokenize(message) 
        message = [word for word in word_tokens if not word in stop_words]
        message = " ".join(message)

        
        #Replace emoji with meaning
        splitted_word = message.split()
        message = [emoji[x] if x in emoji else x for x in splitted_word]
        message = " ".join(message)
        
        #Remove words that contain digits
        splitted_word = message.split()
        message = [word for word in splitted_word if not any(c.isdigit() for c in word)]
        message = " ".join(message)
        
        #Replace non_aphabets with space
        message = re.sub('[^a-zA-Z0-9]',' ', message)
        
        #Removing words with length <=3
        shortword = re.compile(r'\W*\b\w{1,3}\b')
        message=shortword.sub('',message)
        
        #correct all multiple white spaces to a single white space
        message = re.sub('[\s]+',' ',message)
        
        #removing repeated letters of more than two in a words. e.g greeaaatt -> great
        message=re.sub(r'(.)\1+', r'\1\1',message) 
        
        
        #Lemmatizer and Stemmer
        #porter_stemmer = PorterStemmer()
        wordLemm = WordNetLemmatizer()
        #stem_word=''
        lemmatize_word=''
        
        #Stemmer
        '''for word in message.split():
            word=porter_stemmer.stem(word)
            stem_word=stem_word+(word+' ')
            message=stem_word'''

        #Lemmatizer    
        for word in message.split():
            word=wordLemm.lemmatize(word)
            lemmatize_word=lemmatize_word+(word+' ')
            message=lemmatize_word

        results.append(message)
        results_df=pd.DataFrame(results)
 
    return results_df
    
```

```python
train_df['tweet_message']=data_preprocessing(train_df['message'])
test_df['tweet_message']=data_preprocessing(test_df['message'])
```

```python
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_validate = vectorizer.transform(X_validate)
X_test = vectorizer.transform(X_test)
```

### 5.2 Preparing data for modeling(Creating own features)

***5.2.1 Creating own features for train data***

```python
#Character length
data=train_df['tweet_message']
char_len=data.str.split()
char_len=char_len.apply(lambda x : [len(i) for i in x])
char_len=[sum(i) for i in char_len]
char_len=pd.DataFrame(char_len,columns=['Char Length'])

#Number of words
No_of_words=[len(i) for i in data.str.split()]
No_of_words=pd.DataFrame(No_of_words,columns=['No of Words'])

#Part of speech
Part_of_Speech=[]
filtered_POS=[]
for word in data:
    tokens=word.split()
    tags = nltk.pos_tag(tokens)
    counts = Counter(tag for word,tag in tags)
    Part_of_Speech.append(counts)
    
#Filtering out the part of speech we want
for item in Part_of_Speech:
#Part of speech that we want
    pos=['NN','NNP','NNS','JJ','JJR','JJS','RB','RBR','RBS','VB','VBD','VBG','VBN','VBP','VBZ']
    output={x: count for x, count in item.items() if x in pos }
    filtered_POS.append(output)
    
    
#dataframe of features we want
combined_df_test=pd.concat([char_len,No_of_words,pd.DataFrame(filtered_POS).fillna(0)],axis=1)
combined_df_test['Adjective_perc']=(combined_df_test['JJ']+combined_df_test['JJS']+combined_df_test['JJR'])/combined_df_test['No of Words']
combined_df_test['Noun_Perc']=(combined_df_test['NN']+combined_df_test['NNP']+combined_df_test['NNS'])/combined_df_test['No of Words']
combined_df_test['Verb_Perc']=(combined_df_test['VBN']+combined_df_test['VBG']+combined_df_test['VBD']+combined_df_test['VBZ']
             +combined_df_test['VBP']+combined_df_test['VB'])/combined_df_test['No of Words']
combined_df_test['Adverb_Perc']=(combined_df_test['RB']+combined_df_test['RBR']+combined_df_test['RBS'])/combined_df_test['No of Words']
combined_df_test=combined_df_test.drop(['NN','NNP','NNS','JJ','JJR','JJS','RB','RBR','RBS','VB','VBD','VBG','VBN','VBP','VBZ'],axis=1)

#Now using vader to predict sentiment for train
sid_obj = SentimentIntensityAnalyzer()
vader=[]

for item in train_df['tweet_message']:
    sentiment_dict = sid_obj.polarity_scores(item)
    vader.append(sentiment_dict)
vader_train=pd.DataFrame(vader)

#Dataframe of own features
Extra_features_trained=pd.concat([combined_df_test,vader_train,train_df['sentiment']],axis=1)
```

***5.2.2 Creating own features for test data***

```python
#Character length
data=test_df['tweet_message']
char_len=data.str.split()
char_len=char_len.apply(lambda x : [len(i) for i in x])
char_len=[sum(i) for i in char_len]
char_len=pd.DataFrame(char_len,columns=['Char Length'])

#Number of words
No_of_words=[len(i) for i in data.str.split()]
No_of_words=pd.DataFrame(No_of_words,columns=['No of Words'])

#Part of speech
Part_of_Speech=[]
filtered_POS=[]
for word in data:
    tokens=word.split()
    tags = nltk.pos_tag(tokens)
    counts = Counter(tag for word,tag in tags)
    Part_of_Speech.append(counts)
    
#Filtering out the part of speech we want
for item in Part_of_Speech:
#Part of speech that we want
    pos=['NN','NNP','NNS','JJ','JJR','JJS','RB','RBR','RBS','VB','VBD','VBG','VBN','VBP','VBZ']
    output={x: count for x, count in item.items() if x in pos }
    filtered_POS.append(output)

#dataframe for features on test data
combined_df_test=pd.concat([char_len,No_of_words,pd.DataFrame(filtered_POS).fillna(0)],axis=1)
combined_df_test['Adjective_perc']=(combined_df_test['JJ']+combined_df_test['JJS']+combined_df_test['JJR'])/combined_df_test['No of Words']
combined_df_test['Noun_Perc']=(combined_df_test['NN']+combined_df_test['NNP']+combined_df_test['NNS'])/combined_df_test['No of Words']
combined_df_test['Verb_Perc']=(combined_df_test['VBN']+combined_df_test['VBG']+combined_df_test['VBD']+combined_df_test['VBZ']
             +combined_df_test['VBP']+combined_df_test['VB'])/combined_df_test['No of Words']
combined_df_test['Adverb_Perc']=(combined_df_test['RB']+combined_df_test['RBR']+combined_df_test['RBS'])/combined_df_test['No of Words']
combined_df_test=combined_df_test.drop(['NN','NNP','NNS','JJ','JJR','JJS','RB','RBR','RBS','VB','VBD','VBG','VBN','VBP','VBZ'],axis=1)

#Vader for test

vader_test=[]
for item2 in test_df['tweet_message']:
    sentiment_dict2 = sid_obj.polarity_scores(item2)
    vader_test.append(sentiment_dict2)
vader_test=pd.DataFrame(vader_test)

#Dataframe of featues + vader on test
Extra_features_test=pd.concat([combined_df_test,vader_test],axis=1)
```
