```python
def extract_hashtags(df):
    # your code here
    g = df['message'].str.findall(r'@.*?(?=\s|$)')
    df['@mentions'] = g.apply(lambda x: ','.join(x).lower().split(',') if len(x) > 0 else np.nan)
    df["message"]= df['message'].apply(lambda x: ''.join(x) if len(x) > 0 else np.nan)
    df['hashtags'] = df['message'].str.findall(r'#.*?(?=\s|$)')
    df["hashtags"]= df['hashtags'].apply(lambda x: ','.join(x).lower().split(',') if len(x) > 0 else np.nan)
    
    return df
```

```python
def clean_text(text):
    import string
    text.lower()
    text = re.sub(r'@[A-Za-z0-9]+','',text)
    text = re.sub(r'https?://[A-Za-z0-9./]+','',text)
    text = re.sub(r'#[A-Za-z0-9]+','',text)
    
    text = re.sub('\[.*?\]','',text)
    text = re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text = re.sub('\w*\d\w*','',text)
    text = re.sub(r'[^a-zA-Z0-9 ]','',text)
    
    return text
```

```python
df['message'] = df['message'].apply(lambda x: clean_text(x))
```

```python
df.head()
```

```python
def clean_text2(text):
    text = re.sub('[''""...]','',text)
    text = re.sub('\n','',text)
    return text
```

```python
df.message = df.message.apply(lambda x: clean_text2(x))
df.message = df.message.apply(lambda x: x.lower())
```

```python
X = df['message']
y = df['sentiment']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
from sklearn.pipeline  import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
```

```python
textclf = Pipeline([('tfidf', TfidfVectorizer()),('clf', LinearSVC())])
textclf.fit(X_train,y_train)
```

```python
pred = textclf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
print(accuracy_score(y_test,pred))
```
